### Compare the positivity rates/Death rates Pre and Post vaccination

```python

# Since we know that vaccination began on 14th December 2020, we will create 2 dataframes for each period 
pre_vaccine_df = USAdf[(USAdf['Day'] < '2020-12-30')]
pre_vaccine_df.sort_values("Day", ascending= True)
pre_vaccine_df.drop(columns = ['Recovered','Unnamed: 0'])

post_vaccine_df = USAdf[(USAdf['Day'] >= '2020-12-30')]
post_vaccine_df.sort_values("Day", ascending= True)
post_vaccine_df.drop(columns = ['Recovered','Unnamed: 0'])
```

### Average positivity rate pre-vaccination

```python

# Using the dataset created above for pre vaccine, calculate the average positivity rates Pre vaccination

avgprprevaccine = pre_vaccine_df['short_term_positivity_rate'].mean()
print ("The Average positivity rates Pre Vaccination is  : ", avgprprevaccine)
```

### Average positivity rate post-vaccination

```python

# Using the dataset created above for post vaccination period, calculate the average positivity rates post vaccination 

avgprpostvaccine = post_vaccine_df['short_term_positivity_rate'].mean().round(2)
print ("The average positivity rate post vaccination : ", avgprpostvaccine)
```

##### In conclusion, from the data we can see that vaccination had a very slight effect on the positivity rates. Average positivity rate pre-vaccination was 12.12 and 6.75 post vaccination  

### Average Death rates pre-vaccination

```python
# Average positivity rate pre-vaccine
# Using the dataset created above for pre vaccine, calculate the average Death rates pre vaccination

avgprevacdeaths = pre_vaccine_df['Deaths'].mean()
print ("The average Death rates Pre vaccination : ", avgprevacdeaths)
```

### Average Death rates post-vaccination

```python
# Using the dataset created above for post vaccination period, calculate the average Death rates post vaccination

avgpostvacdeaths = post_vaccine_df['Deaths'].mean().round(2)
print ("The average Death rates post- vaccination is : ", avgpostvacdeaths)
```

##### In conclusion, Average of 331702.2  deaths pre-vaccination and 507589.86 deaths post vaccination, showing a steady rise in the average number of deaths regardless of vaccination.

### Find the total number of infections during the month that recorded the highst number of vaccination (April 2021)

```python

# April of 2021 had the highest number of vaccinations
# We need to know the first day of the month and the last day of the month
# set the desired start date and end date to filter *df*
start_date = '2021-04-01'
end_date = '2021-04-30'

# set the df5
df5 = (USAdf['Day'] > start_date) & (USAdf['Day'] <= end_date)

# assign df5 to dfr to return the rows with days between our specified start/end dates
dfr = USAdf.loc[df5]

# count the total number of vaccinations during the month of april 2021
Total_vaccinations = dfr['people_vaccinated'].mean().round(2)
print ("The total number of vaccinations in the moth of April 2021 is : ", Total_vaccinations)
```
